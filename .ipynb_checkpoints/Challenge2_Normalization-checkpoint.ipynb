```python

def getRed(redVal):
    return '#%02x%02x%02x' % (redVal, 0, 0)
def getGreen(greenVal):
    return '#%02x%02x%02x' % (0, greenVal, 0)
def getBlue(blueVal):
    return '#%02x%02x%02x' % (0, 0, blueVal)

def traceHistogram(im):
    list_img=im.histogram()

    l1 = list_img[0:256]
    # Take only the Blue counts
    l2 = list_img[256:512]
    # Take only the Green counts
    l3 = list_img[512:768]
    plt.figure(0)

    #RED
    for i in range(0, 256):
        plt.bar(i, l1[i], color = getRed(i), edgecolor=getRed(i), alpha=0.3)
    # GREEN
    plt.figure(1)
    for i in range(0, 256):
        plt.bar(i, l2[i], color = getGreen(i), edgecolor=getGreen(i),alpha=0.3)

    # B histogram
    plt.figure(2)

    for i in range(0, 256):
        plt.bar(i, l3[i], color = getBlue(i), edgecolor=getBlue(i),alpha=0.3)

    plt.show()
```

```python
#from matplotlib.pyplot import imshow

#%matplotlib inline
#imshow(np.asarray(pil_im))
new_im=make_square(pil_im)
display(new_im)
traceHistogram(new_im)
```

```python
autocontrast_im=ImageOps.equalize(new_im)
traceHistogram(autocontrast_im)
```

```python
#PREDICTIONS

#Loop on locally downloaded images
from os import listdir
from os.path import isfile, join, isdir

input_path = 'gear_images'
output_path = 'training_data'

labels = [f for f in listdir(input_path) if isdir(join(input_path, f))]
for label in labels:
    folder = input_path + '/' + label
    files = [f for f in listdir(folder) if isfile(join(folder, f))]
    i=0
    for file in files:
        img=Image.open(folder+'/'+file, 'r')
        img=make_square(img)
        img=ImageOps.equalize(img)
        #img.save(np.asarray(img)) 
        with open(output_path + '/' + label + '_' + str(i) + '.dat', 'wb') as of:
            of.write(np.asarray(img))
        i += 1
        # DO stuff
    
```
